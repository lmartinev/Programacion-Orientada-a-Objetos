# while

Dado un capital inicial, si se paga una tasa de interés del 3% anual despliega cuantos años se tardará en reunir la cantidad necesaria para comprar un coche (dada por el usuario)

```python
cantidad=float(input("cuanto cuesta el coche que quieres comprar? "))
capital=float(input("Que capital tienes? "))
años=0
while capital<cantidad:
   años+=1       #años=años+1
   capital*=1.03  #capital=capital* 1.03
   print(capital)
print(años)
```

```python
#swap
a,b=0,1  #asigna simultaneamente a=0 y b=1
a,b=b,a  #intercambia
print(a)
print(b)
```

 **Serie de Fibonacci**

```python
a,b=0,1 
while a<10:
    print(a,end=",")
    a,b=b,a+b
```

```python
#suma numeros del 20 al 30
i,suma=20,0
while i<30:
    suma+=i
    i=i+1
else:    
    print(suma)
```

**Validaciones**

*Ejemplo uso de break*

```python
#vaidacion de una entrada entre 1 y 10 inclusive
while True:
    num=int(input("Numero: "))
    if num>=1 and num<=10:
        break #sale del ciclo
    else:
        print("numero invalido")
print("finalmente dio un numero valido")    
```

```python
r=range(10)
inrange=False
while not inrange:
    num=int(input("Numero: "))
    if num in r:
        inrange=True
else:
    print("Valido ",num)
print("salio")
```

*Ejemplo uso de continue*

```python
var = 10                    
while var > 0:              
    var = var -1
    if var == 5:
        continue  
    print(var,end=',') 
print("Adios") 
```

## Listas  
- secuencias, ordenadas, que permiten cambios y duplicados

```python
lista=[2,4,6,8]
for n in lista:           
    print("numero ", n)
```

```python
colores=['rojo','azul','amarillo']
for c in colores:           
    print("color", c)
```

```python
print(colores)
print(len(colores))  #tamaño de la lista
```

```python
#añadir a la lista
colores.append('verde')
colores
```

```python
#insertar en medio 
colores.insert(2,'naranja')
print(colores)
```

```python
colores.sort()
print(colores)
```

```python
print(colores[1])
```
